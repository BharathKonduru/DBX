###Fact_Order  

```python
%sql
CREATE TABLE IF NOT EXISTS dbx.ecommerce.fact_order (
    order_id    INT PRIMARY KEY,
    customer_id INT REFERENCES dbx.ecommerce.dim_customer(customer_id),
    date_id     INT REFERENCES dbx.ecommerce.dim_date(date_id),
    product_id  INT REFERENCES dbx.ecommerce.dim_product(product_id),
    amount      DECIMAL(12, 2)
);
```

```python
%sql
INSERT INTO dbx.ecommerce.fact_order (order_id, customer_id, date_id, product_id, amount) VALUES
(1001, 1, 1, 1, 115.00),
(1002, 2, 2, 2,  42.00),
(1003, 3, 3, 3, 110.00),
(1004, 4, 4, 4, 395.00),
(1005, 5, 5, 5,  48.00);
```

###Fact_Order_Item  

```python
%sql
CREATE TABLE IF NOT EXISTS dbx.ecommerce.fact_order_item (
    order_item_id   INT PRIMARY KEY,
    order_id        INT REFERENCES dbx.ecommerce.fact_order(order_id),
    product_id      INT REFERENCES dbx.ecommerce.dim_product(product_id),
    quantity        INT
);
```
