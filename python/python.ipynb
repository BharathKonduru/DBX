### List Comprehension

Traditional approach

```python
squares = []
for i in range(5):
    squares.append(i * i)

print(squares)
# [0, 1, 4, 9, 16]
```

List comprehension

```python
squares = [i * i for i in range(5)]
print(squares)
# [0, 1, 4, 9, 16]
```

List comprehension - With Condition

```python
evens = [i for i in range(10) if i % 2 == 0]
print(evens)
# [0, 2, 4, 6, 8]
```

List comprehension - Transforming Data

```python
names = ["energy", "data", "python"]

upper_names = [name.upper() for name in names]
print(upper_names)
# ['ENERGY', 'DATA', 'PYTHON']
```

List comprehension - Practical (Data Engineering context)

```python
mw_values = [1.2, 0.8, 3.5]

kw_values = [mw * 1000 for mw in mw_values]
print(kw_values)
# [1200, 800, 3500]
```

List comprehension - Nested List Comprehension

```python
matrix = [[1, 2], [3, 4], [5, 6]]

flat = [num for row in matrix for num in row]
print(flat)
# [1, 2, 3, 4, 5, 6]
```

###Find Duplicates in a List — Write a function to find duplicates in a list and return them as a set

```python
def find_duplicates(lst):
    seen = set()
    duplicates = set()
    
    for item in lst:
        if item in seen:
            duplicates.add(item)
        else:
            seen.add(item)
    
    return duplicates

# Example
numbers = [1, 2, 3, 4, 2, 5, 1, 6]
print(find_duplicates(numbers))
# Output: {1, 2}
```

###Implement a Stack — Implement a stack data structure in Python.

A stack is a linear data structure that follows LIFO (Last In, First Out) — the last element pushed is the first one popped.

```python
class Stack:
    def __init__(self):
        self.items = []

    # Push element onto stack
    def push(self, item):
        self.items.append(item)

    # Pop element from stack
    def pop(self):
        if self.is_empty():
            return "Stack is empty"
        return self.items.pop()

    # Peek top element
    def peek(self):
        if self.is_empty():
            return "Stack is empty"
        return self.items[-1]

    # Check if stack is empty
    def is_empty(self):
        return len(self.items) == 0

    # Get stack size
    def size(self):
        return len(self.items)

    # Display stack
    def display(self):
        return self.items

s = Stack()

s.push(10)
s.push(20)
s.push(30)

print("Stack:", s.display())
print("Top element:", s.peek())

print("Popped:", s.pop())
print("After pop:", s.display())

print("Is empty?", s.is_empty())
print("Size:", s.size()) 
```

###	Check Palindrome — Write a function to check if a given string is a palindrome.


```python
def is_palindrome(s):
    s = s.lower().replace(" ", "")
    left, right = 0, len(s) - 1

    while left < right:
        if s[left] != s[right]:
            return False
        left += 1
        right -= 1

    return True

print(is_palindrome("A man, a plan, a canal: Panama"))  # True
print(is_palindrome("No lemon, no melon"))              # True
print(is_palindrome("Hello World"))                     # False
```

###Find Missing Number — Given a list of integers from 1 to n, with one number missing, write a function to find the missing number.

```python
def find_missing(nums):
    n = len(nums) + 1
    expected_sum = n * (n + 1) // 2
    actual_sum = sum(nums)
    return expected_sum - actual_sum

nums = [1, 2, 4, 5]
print(find_missing(nums))   # 3
```

###Count Vowels and Consonants — Write a function to count the number of vowels and consonants in a given string.

```python
def count_vowels_consonants(s):
    vowels = "aeiou"
    vowel_count = 0
    consonant_count = 0

    for char in s.lower():
        if char.isalpha():          # consider only letters
            if char in vowels:
                vowel_count += 1
            else:
                consonant_count += 1

    return vowel_count, consonant_count

text = "Hello World"
vowels, consonants = count_vowels_consonants(text)

print("Vowels:", vowels)        # 3
print("Consonants:", consonants)  # 7
```

###Merge Sorted Lists — Write a function to merge two sorted lists into a single sorted list.

```python
def merge_sorted_lists(list1, list2):
    i, j = 0, 0
    merged = []

    while i < len(list1) and j < len(list2):
        if list1[i] <= list2[j]:
            merged.append(list1[i])
            i += 1
        else:
            merged.append(list2[j])
            j += 1

    # Add remaining elements
    merged.extend(list1[i:])
    merged.extend(list2[j:])

    return merged

l1 = [1, 3, 5, 7]
l2 = [2, 4, 6, 8]

print(merge_sorted_lists(l1, l2))
```

###Implement a Queue — Implement a queue data structure in Python.

A queue removes elements in the order they were inserted.

```python
class Queue:
    def __init__(self):
        self.items = []

    # Enqueue (add to rear)
    def enqueue(self, item):
        self.items.append(item)

    # Dequeue (remove from front)
    def dequeue(self):
        if self.is_empty():
            return "Queue is empty"
        return self.items.pop(0)

    # Peek front element
    def peek(self):
        if self.is_empty():
            return "Queue is empty"
        return self.items[0]

    # Check if empty
    def is_empty(self):
        return len(self.items) == 0

    # Get size
    def size(self):
        return len(self.items)

    # Display queue
    def display(self):
        return self.items
    
q = Queue()

q.enqueue(10)
q.enqueue(20)
q.enqueue(30)

print("Queue:", q.display())
print("Front:", q.peek())

print("Dequeued:", q.dequeue())
print("After dequeue:", q.display())
```

###Find Maximum Subarray Sum — Write a function to find the maximum sum of a contiguous subarray within a one-dimensional array of numbers.

```python
def max_subarray_sum(nums):
    max_sum = nums[0]
    current_sum = nums[0]

    for num in nums[1:]:
        current_sum = max(num, current_sum + num)
        max_sum = max(max_sum, current_sum)

    return max_sum

arr = [-2, 1, -3, 4, -1, 2, 1, -5, 4]
print(max_subarray_sum(arr))
```

###How to find the maximum value in a list without using the max function?

```python
def find_max(nums):
    maximum = nums[0]          # assume first element is max

    for num in nums:
        if num > maximum:
            maximum = num

    return maximum

numbers = [3, 7, 2, 9, 5]
print(find_max(numbers))   # 9
```

###Write a program to check if a number is a prime number.

```python
def is_prime(n):
    if n <= 1:
        return False

    for i in range(2, n):
        if n % i == 0:
            return False

    return True

print(is_prime(7))   # True
print(is_prime(10))  # False
```

###How to find unique and duplicate elements from a list?

```python
def find_unique_duplicates(nums):
    seen = set()
    duplicates = set()

    for num in nums:
        if num in seen:
            duplicates.add(num)
        else:
            seen.add(num)

    unique = seen - duplicates
    return list(unique), list(duplicates)

nums = [1, 2, 3, 2, 4, 5, 1, 6]
unique, duplicates = find_unique_duplicates(nums)
print("Unique:", unique)         # [3,4,5,6]
print("Duplicates:", duplicates) # [1,2]
```
